```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
data = pd.read_csv("insurance.csv")
data.head()
```

```python
#Plot a scatter plot with age on the x-axis and charges on the y-axis.  
x = data.iloc[:,:1].values
y = data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
#Using linear_model.LinearRegression() from sklearn, fit a model to your data
# Fit the model
charges_model = LinearRegression()
charges_model.fit(x,y)
```

```python
# make predictions on data
unk_x = [[4]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

print("Insurance charges should be:", charges_model.predict(unk_x))
```

```python
#Plot another scatter plot with the best-fit line.
# Plot the data and model
y_pred = charges_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```
